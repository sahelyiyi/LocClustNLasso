At this notebook, I have tested the regression algorithm for SBM graph with 2 clusters, each contains 150 nodes, and also pin=0.7 and normal distribution with mean 8, pout=0.01 and normal distribution with mean 2.
X features are selected from 2d Gaussian Distribution with the dimensions 2*5 and W1 = [10, 15] and W2 = [2, 4]. also sampling set ratio is 0.3 . The number of iterations is 300 for each test.


The following charts are the mse during iterations.

```python
import matplotlib.pyplot as plt
from reg_sbm import *

our_mses = []
simple_regression_mses = []
for i in range(10):
    iteration_mses, mse = run()
    our_mses.append(iteration_mses[-1])
    simple_regression_mses.append(mse)
    
    indices = [i for i in range(len(iteration_mses))]
    plt.plot(indices, iteration_mses)
#     plt.ylim(-.1, 1.5)
    plt.title('mse curve during iterations for test number %d' % i)
    plt.xlabel('Iteration')
    plt.ylabel('mse')
    plt.show()
```

The following chart shows the difference between our method and simple regression method for these 10 tests.

```python
indices = [i for i in range(len(simple_regression_mses))]
plt.plot(indices, our_mses, label = "our method")
plt.ylim(0, 3)
plt.plot(indices, simple_regression_mses, label = "simple regression")
plt.title('our mse vs simple regression mse for tests')
plt.xlabel('Iteration')
plt.ylabel('mse')
plt.show()
```

```python
print('mean of our mse:', np.mean(our_mses))
print('mean of simple regression mse:', np.mean(simple_regression_mses))
```
