At this notebook, I have tested the regression algorithm for SBM graph with 2 clusters, each contains 150 nodes, and also pin=0.5 with normal distribution of mean 8 for the edge weights, pout=0.01 with normal distribution of mean 2 for the edge weights.
X features are selected from 2d Gaussian Distribution with the dimensions 2*5 and W1 = [0.5, 0.7] and W2 = [2, 4]. also sampling set ratio is 0.3 . The number of iterations is 300 for each test.


The following charts are the mean_squared_error during iterations.

```python
import matplotlib.pyplot as plt
from reg_sbm import *

our_mses = []
simple_regression_mses = []
decision_tree_mses = []
for i in range(5):
    iteration_mses, regression_mse, decision_tree_mse  = run_reg_sbm()
    our_mses.append(iteration_mses[-1])
    simple_regression_mses.append(regression_mse)
    decision_tree_mses.append(decision_tree_mse)
    
    if i == 0:
        indices = [i for i in range(len(iteration_mses))]
        plt.plot(indices, iteration_mses)
    #     plt.ylim(-.1, 1.5)
        plt.title('mse curve during iterations for test number %d' % i)
        plt.xlabel('Iteration')
        plt.ylabel('mse')
        plt.show()
```

The following chart shows the difference between our method and simple regression method for these 5 tests.

```python
print('mean of our mean_squared_error:', np.mean(our_mses))
print('mean of simple regression mean_squared_error:', np.mean(simple_regression_mses))
print('mean of decision tree regression mean_squared_error:', np.mean(decision_tree_mses))
```

```python
indices = [i for i in range(len(simple_regression_mses))]
plt.plot(indices, our_mses, label = "our method")
# plt.ylim(0, 3)
plt.plot(indices, simple_regression_mses, label = "simple regression")
plt.plot(indices, decision_tree_mses, label = "decision tree regression")
plt.title('our mse vs other methods mses')
plt.xlabel('Iteration')
plt.ylabel('mse')
plt.yscale('log')
plt.show()
```
